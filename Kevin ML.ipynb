```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as stats
import sklearn as sklearn
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, accuracy_score




from sklearn import linear_model



data = pd.read_csv('diabetes_binary_health_indicators_BRFSS2015.csv')
```

```python
data['Obese'] = np.where(data['BMI'] >= 30, 1, 0)

data.head()
row_count = data.shape[0]
print("Number of rows:", row_count)
```

Next we must split the data into training and a test set. 80 % of the data will be randomly chosen to be in the training set, the rest of the data will be for the test set. 

```python
count=int(len(data)*0.8)
training=data.sample(count)
target=training['Obese']
training=training.drop(columns=['Obese'])
training=training.drop(columns=['BMI'])

training.head()




```

```python
test=data.drop(training.index)
obese=test['Obese']
test=test.drop(columns=['Obese'])
test=test.drop(columns=['BMI'])


test.head()
print(obese)
```

```python
lr=linear_model.LogisticRegression(max_iter=1000,class_weight='balanced')
lr.fit(training,target)
```

We will fit a logistic regression model fro the training set, using the Obese variable as the target

```python
preds=lr.predict(test)
print(preds)

correctPred=(preds==obese)
res=correctPred.sum()/len(obese)
print(res)
```

An SVM model could work?

```python

trainingSVM, testSVM = train_test_split(data, test_size=0.2, stratify=data['Obese'], random_state=125)

target_train = trainingSVM['Obese']
target_test = testSVM['Obese']
trainingSVM = trainingSVM.drop(columns=['Obese', 'BMI'])
testSVM = testSVM.drop(columns=['Obese', 'BMI'])
```

```python
scaler = StandardScaler()
trainingScaler = scaler.fit_transform(trainingSVM)
testScaler = scaler.transform(testSVM)
```

```python
svm=SVC(kernel='linear',C=1,gamma='scale',class_weight='balanced',max_iter=5000)
svm.fit(trainingScaler,target_train)
```

```python

SVMpred = svm.predict(testScaler)
print("Accuracy:", accuracy_score(target_test, SVMpred))

```

We had to add a max iteration since the model was too big to create a SVM otherwise, another way to do this could be to get a random subset of the dataset to create the model off. 

```python
subset = data.sample(n=10000, random_state=125)

trainingSVM2, testSVM2 = train_test_split(subset, test_size=0.2, stratify=subset['Obese'], random_state=125)

target_train2 = trainingSVM2['Obese']
target_test2 = testSVM2['Obese']
trainingSVM2 = trainingSVM2.drop(columns=['Obese', 'BMI'])
testSVM2 = testSVM2.drop(columns=['Obese', 'BMI'])
```

```python
scaler2 = StandardScaler()
trainingScaler2 = scaler2.fit_transform(trainingSVM2)
testScaler2 = scaler2.transform(testSVM2)
```

```python
C_values = [0.1, 1, 5, 10, 50, 100]
best_C = None
best_accuracy = 0

for C in C_values:
    svm = SVC(kernel='linear', C=C, gamma='scale', class_weight='balanced')
    svm.fit(trainingScaler2, target_train2)
    preds = svm.predict(testScaler2)
    accuracy = accuracy_score(target_test2, preds)
    print(f"C={C}, Accuracy={accuracy}")
    
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_C = C

```

```python
svm2=SVC(kernel='linear',C=1,gamma='scale',class_weight='balanced')
svm2.fit(trainingScaler2,target_train2)
```

```python
SVMpred2 = svm2.predict(testScaler2)
print("Accuracy:", accuracy_score(target_test2, SVMpred2))
```

```python
df_latin = pd.read_csv('latin_america_obesity.csv')
display(df_latin.head())
print(df_latin.dtypes)

# Many of the columns in this dataset do not have very meaningful column names, so we are renaming them
df_latin.rename(columns = {'FAVC' : 'eat_high_caloric', 'FCVC' : 'eat_vegetables', 'NCP' : 'num_meals_per_day', 'CAEC' : 'food_between_meals', 'CH2O' : 'water_per_day', 'FAF' : 'physical_activity', 
                           'TUE' : 'technology_usage', 'CALC' : 'alcohol_consumption', 'MTRANS' : 'transportation_method', 'SCC' : 'calories_per_day'}, inplace = True)
df_latin.head()


# Here, we encode the categorical features with numeric integer values to indicate their levels.
df_latin['Gender'] = df_latin['Gender'].replace({'Female' : 0, 'Male' : 1})
df_latin['family_history_with_overweight'] = df_latin['family_history_with_overweight'].replace({'yes' : 1, 'no' : 0}).astype(int)
df_latin['eat_high_caloric'] = df_latin['eat_high_caloric'].replace({'yes' : 1, 'no' : 0}).astype(int)
df_latin['SMOKE'] = df_latin['SMOKE'].replace({'yes' : 1, 'no' : 0}).astype(int)
df_latin['calories_per_day'] = df_latin['calories_per_day'].replace({'yes' : 1, 'no' : 0}).astype(int)
df_latin['food_between_meals'] = df_latin['food_between_meals'].replace({'no' : 0, 'Sometimes' : 1, 'Frequently' : 2, 'Always' : 3}).astype(int)
df_latin['alcohol_consumption'] = df_latin['alcohol_consumption'].replace({'no' : 0, 'Sometimes' : 1, 'Frequently' : 2, 'Always' : 3}).astype(int)
df_latin['transportation_method'] = df_latin['transportation_method'].replace({'Bike' : 1, 'Motorbike' : 2, 'Walking' : 3, 'Automobile' : 4, 'Public_Transportation' : 5})


'''For this dataset, we do not have features for BMI or Obesity. We will need to create them manually, 
   so create BMI from its definition of weight/(height^2) and create Obese feature from the definition that an Obese individual has a BMI of at least 30'''
df_latin['BMI'] = round(df_latin['Weight'] / (df_latin['Height'] ** 2), 1)
df_latin['Obese'] = np.where(df_latin['BMI'] >= 30, 1, 0)

'''remove the previous obesity column since we create the binary obesity column based off the value of BMI. 
   This is to ensure consistency in how we determine the obesity feature between the two datasets.'''
df_latin = df_latin.drop('NObeyesdad', axis = 1) 
display(df_latin.head())

X_latin = df_latin.drop(columns = ['BMI', 'Obese'])
y_latin = df_latin[['Obese']]

X_train_latin, X_test_latin, y_train_latin, y_test_latin = train_test_split(X_latin, y_latin, test_size = 0.25, random_state = 125)
```
