```python
import pandas as pd
from IPython.display import display

main_df = pd.read_csv('diabetes_binary_health_indicators_BRFSS2015.csv')
display(main_df.head(10))
```

```python
second_df = pd.read_csv('latin_america_obesity.csv')
display(second_df.head(10))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

plt.figure(figsize=(8, 6))
sns.countplot(data=second_df, x='Gender', palette='pastel')


plt.title('Gender Distribution in Secondary Dataset')
plt.xlabel('Gender')
plt.ylabel('Count')

plt.show()
```

```python
age_summary = second_df['Age'].describe(percentiles=[0.25, 0.5, 0.75])

min_age = age_summary['min']
q1_age = age_summary['25%']
median_age = age_summary['50%']
q3_age = age_summary['75%']
max_age = age_summary['max']

five_number_summary = {
    'Minimum': min_age,
    'Q1 (25%)': q1_age,
    'Median (50%)': median_age,
    'Q3 (75%)': q3_age,
    'Maximum': max_age
}

five_number_summary
```

```python
age_bins = range(10, 61, 10)
age_labels = [f"{i}-{i+9}" for i in age_bins[:-1]]


second_df['Age_Group'] = pd.cut(second_df['Age'], bins=age_bins, labels=age_labels, right=False)

plt.figure(figsize=(10, 6))
sns.countplot(data=second_df, x='Age_Group', palette='pastel')


plt.title('Age Distribution in Second Dataset')
plt.xlabel('Age Range')
plt.ylabel('Count')

plt.xticks(rotation=45)

plt.show()
```
