### **PART 3: VISUALIZATION**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load the Latin American dataset

```python
latin_america_data = pd.read_csv('latin_america_obesity.csv')
```

1. BMI for Latin America Datasets 
2. common age bins + labels to compare

```python
latin_america_data['Calculated_BMI'] = latin_america_data['Weight'] / (latin_america_data['Height'] ** 2)
age_bins = [18, 24, 29, 34, 39, 44, 49, 54, 59, 64, 69, 74, 79, float('inf')]
age_labels = ['18-24', '25-29', '30-34', '35-39', '40-44', '45-49', '50-54', '55-59', '60-64', '65-69', '70-74', '75-79', '80+']
```

```python
latin_america_data['Age_Range'] = pd.cut(latin_america_data['Age'], bins=age_bins, labels=age_labels)
```

BMI Distribution Plot for Latin America

```python
plt.figure(figsize=(16, 10))
sns.boxplot(
    data=latin_america_data, 
    x='Age_Range', 
    y='Calculated_BMI', 
    color='#CA9BF7',
    width=0.6, 
    linewidth=2, 
    fliersize=8, 
    flierprops=dict(marker='o', color='coral', alpha=0.6)
)

# Customize titles and labels for a friendly look
plt.title('BMI Distribution Across Age Groups\n(Latin American Dataset)', fontsize=20, fontweight='bold', color='teal')
plt.xlabel('Age Group', fontsize=16, color='dimgray')
plt.ylabel('BMI', fontsize=16, color='dimgray')

# Rotate x-axis labels for readability
plt.xticks(rotation=45, ha='right', fontsize=14)
plt.yticks(fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.4)

# Display without legend for a cleaner look
plt.tight_layout(pad=2)
plt.show()
```
