```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as stats
import sklearn as sklearn
from sklearn import linear_model



data = pd.read_csv('diabetes_binary_health_indicators_BRFSS2015.csv')
```

```python
data['Obese'] = np.where(data['BMI'] >= 30, 1, 0)

data.head()
```

```python
data.describe()
```

```python
zero = (data == 0).sum()
zero
```

Next we must split the data into training and a test set. 80 % of the data will be randomly chosen to be in the training set, the rest of the data will be for the test set. 

```python
count=int(len(data)*0.8)
training=data.sample(count)
target=training['Obese']
training=training.drop(columns=['Obese'])
training=training.drop(columns=['BMI'])


training.head()
```

```python
test=data.drop(training.index)
obese=test['Obese']
test=test.drop(columns=['Obese'])
test=test.drop(columns=['BMI'])


test.head()
print(obese)
```

```python
lr=linear_model.LogisticRegression(max_iter=1000)
lr.fit(training,target)
```

We will fit a logistic regression model fro the training set, using the Obese variable as the target

```python
preds=lr.predict(test)
print(preds)

correctPred=(preds==obese)
res=correctPred.sum()/len(obese)
print(res)
```

The model has an accuracy of .69

```python
help=stats.add_constant(training)
model=stats.OLS(target,help).fit()
print(model.summary())
```
